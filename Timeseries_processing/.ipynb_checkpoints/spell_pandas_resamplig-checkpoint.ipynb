```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import datetime as dt
```

```python
rain = np.asarray([0.9, 0.1 , 0, 0.1, 0, 6, 3, 0, 0, 0, 14, 0, 0 ,0, 5,6,4,0, 0, 0, 11, 5, 1, 0, 0, 1, 1, 1, 2, 11, 2, 1, 0])
```

```python
wet_th = 1
```

### Version 1 (my own)

```python
index = pd.date_range('1/1/2000', periods=len(rain), freq='H')
srain = pd.Series(rain, index=index)

srain =srain.to_frame()
srain = srain.rename(columns = {0:'pr'})

# srain = srain.to_dataframe()
# srain = srain.rename(columns = {'RAIN':'pr'})
srain['pr'] = srain['pr'].where(srain['pr']>=wet_th,0)
srain['event_start'] = (srain['pr'].astype(bool).shift() != srain['pr'].astype(bool)).where(srain['pr']>=wet_th, False)
srain['event_end'] = (srain['pr'].astype(bool).shift(-1) != srain['pr'].astype(bool)).where(srain['pr']>=wet_th, False)
srain['event'] = srain['event_start'].cumsum().where(srain['pr']>=wet_th)

wet_event_intensity = srain.groupby('event')['pr'].sum()
wet_event_duration = srain.groupby('event')['pr'].count()
srain['event_pr'] = srain['event'].map(wet_event_intensity).where(srain['event_end'])
srain['event_dur'] =  srain['event'].map(wet_event_duration).where(srain['event_end'])
```

### Version 2 (from the web)

```python
index = pd.date_range('1/1/2000', periods=len(rain), freq='H')
srain = pd.Series(rain, index=index)

srain =srain.to_frame()
srain = srain.rename(columns = {0:'pr'})


srain['pr'] = srain['pr'].where(srain['pr']>=1,0)
srain['event'] = (srain['pr'].astype(bool).shift() != srain['pr'].astype(bool)).cumsum()
srain['event_start'] = (srain['pr'].astype(bool).shift() != srain['pr'].astype(bool))
srain['wet_event'] = ((srain['pr'].astype(bool).shift() != srain['pr'].astype(bool)).where(srain['pr']>=1,0)).cumsum().where(srain['pr']>=1)
srain['wet_event_start'] = (srain['pr'].astype(bool).shift() != srain['pr'].astype(bool)).where(srain['pr']>=1,0)
srain['wet_event_end'] = (srain['pr'].astype(bool).shift(-1) != srain['pr'].astype(bool)).where(srain['pr']>=1,0)
srain = srain.reset_index()
srain['day'] = srain['index'].dt.normalize()

#event duration
wet_event_intensity = srain.groupby('wet_event')['pr'].sum()

# group by day to get unique block count and value count
session_map = srain[srain['pr'].astype(bool)].groupby('day')['event'].nunique()
hour_map = srain[srain['pr'].astype(bool)].groupby('day')['event'].count()

# map to original dataframe
srain['sessions'] = srain['day'].map(session_map)
srain['hours'] = srain['day'].map(hour_map)

# calculate result
res = srain.groupby(['day', 'hours', 'sessions'], as_index=False)['pr'].sum()
res['duration'] = res['hours'] / res['sessions']
res['amount'] = res['pr'] / res['sessions']
```
